```python
# ═══════════════════════════════════════════════════
#  USER CONFIG  — edit these three lines only
# ═══════════════════════════════════════════════════
FILE_PATH   = 'your_dataset.csv'   # Path to your file
TARGET_COL  = 'target'             # Your target / label column
TASK_TYPE   = 'auto'               # 'classification', 'regression', or 'auto'
# ═══════════════════════════════════════════════════

def load_data(path: str) -> pd.DataFrame:
    """Smart loader — detects format from extension."""
    ext = os.path.splitext(path)[-1].lower()
    loaders = {
        '.csv'   : lambda p: pd.read_csv(p),
        '.tsv'   : lambda p: pd.read_csv(p, sep='\t'),
        '.txt'   : lambda p: pd.read_csv(p, sep=None, engine='python'),
        '.xlsx'  : lambda p: pd.read_excel(p, engine='openpyxl'),
        '.xls'   : lambda p: pd.read_excel(p, engine='xlrd'),
        '.json'  : lambda p: pd.read_json(p),
        '.parquet': lambda p: pd.read_parquet(p),
        '.feather': lambda p: pd.read_feather(p),
        '.pkl'   : lambda p: pd.read_pickle(p),
        '.html'  : lambda p: pd.read_html(p)[0],
    }
    if path.startswith('http'):
        if '.csv' in path: return pd.read_csv(path)
        if '.xlsx' in path: return pd.read_excel(path)
        return pd.read_csv(path)
    loader = loaders.get(ext)
    if loader is None:
        raise ValueError(f'Unsupported extension: {ext}')
    return loader(path)

df_raw = load_data(FILE_PATH)
df     = df_raw.copy()
print(f'✅ Loaded  →  {df.shape[0]:,} rows × {df.shape[1]} columns')
df.head()
```

```python
# ── Comprehensive Column Summary ──────────────────────────────────────────────
def column_summary(df: pd.DataFrame) -> pd.DataFrame:
    summary = pd.DataFrame({
        'dtype'     : df.dtypes,
        'non_null'  : df.notnull().sum(),
        'null'      : df.isnull().sum(),
        'null_%'    : (df.isnull().sum() / len(df) * 100).round(2),
        'unique'    : df.nunique(),
        'unique_%'  : (df.nunique() / len(df) * 100).round(2),
        'sample'    : [df[c].dropna().iloc[0] if df[c].notnull().any() else None for c in df.columns],
    })
    return summary.sort_values('null_%', ascending=False)

summary = column_summary(df)
display(summary)
```

```python
# ── Drop high-null columns (threshold configurable) ────────────────────────────
NULL_THRESHOLD = 0.60   # drop columns with > 60% nulls
high_null_cols = [c for c in df.columns if df[c].isnull().mean() > NULL_THRESHOLD]
print(f'Columns with >{NULL_THRESHOLD*100:.0f}% nulls → dropping: {high_null_cols}')
df.drop(columns=high_null_cols, inplace=True, errors='ignore')
```

```python
# ── Fix Data Types ────────────────────────────────────────────────────────────

# Convert object columns that look numeric
for col in df.select_dtypes('object').columns:
    try:
        df[col] = pd.to_numeric(df[col], errors='raise')
    except (ValueError, TypeError):
        pass

# Parse date columns (add known date column names here)
# date_like = ['date', 'timestamp', 'created_at', 'updated_at']
# for col in date_like:
#     if col in df.columns:
#         df[col] = pd.to_datetime(df[col], errors='coerce')

# Convert low-cardinality strings to category
for col in df.select_dtypes('object').columns:
    if df[col].nunique() / len(df) < 0.05:
        df[col] = df[col].astype('category')

# Downcast numeric types to save memory
df = df.apply(lambda c: pd.to_numeric(c, downcast='integer')
              if is_numeric_dtype(c) and (c == c.astype(int)).all()
              else c, axis=0)

print('Data types after cleaning:')
print(df.dtypes.value_counts())
```

```python
# ── Strip whitespace from strings ─────────────────────────────────────────────
str_cols = df.select_dtypes('object').columns
df[str_cols] = df[str_cols].apply(lambda c: c.str.strip())

# ── Standardise column names ──────────────────────────────────────────────────
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r'[\s\-]+', '_', regex=True)
    .str.replace(r'[^\w]', '', regex=True)
)
# Update target/feature lists after renaming
TARGET_COL = TARGET_COL.strip().lower().replace(' ', '_').replace('-', '_')
num_cols   = df.select_dtypes(include=np.number).columns.tolist()
cat_cols   = df.select_dtypes(include=['object', 'category']).columns.tolist()
feat_num   = [c for c in num_cols if c != TARGET_COL]
feat_cat   = [c for c in cat_cols if c != TARGET_COL]

print('Columns after rename:', df.columns.tolist())
```

```python
# ── Handle Outliers — IQR fence & Z-score ─────────────────────────────────────

def detect_outliers_iqr(df: pd.DataFrame, cols: list, factor: float = 1.5) -> pd.DataFrame:
    """Flag outliers per column using IQR."""
    report = []
    for col in cols:
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR    = Q3 - Q1
        lo, hi = Q1 - factor * IQR, Q3 + factor * IQR
        n_out  = ((df[col] < lo) | (df[col] > hi)).sum()
        report.append({'column': col, 'lower': lo, 'upper': hi,
                       'n_outliers': n_out, 'pct': round(n_out / len(df) * 100, 2)})
    return pd.DataFrame(report).sort_values('n_outliers', ascending=False)

def detect_outliers_zscore(df: pd.DataFrame, cols: list, threshold: float = 3.0) -> pd.Series:
    """Boolean mask: True where any column has |z| > threshold."""
    z = np.abs(stats.zscore(df[cols].dropna()))
    return (z > threshold).any(axis=1)

outlier_report = detect_outliers_iqr(df, feat_num)
display(outlier_report)

# ── Cap outliers with Winsorization (alternative to removal) ─────────────────
def winsorize_df(df: pd.DataFrame, cols: list, lower_pct=0.01, upper_pct=0.99):
    df = df.copy()
    for col in cols:
        lo = df[col].quantile(lower_pct)
        hi = df[col].quantile(upper_pct)
        df[col] = df[col].clip(lo, hi)
    return df

# df = winsorize_df(df, feat_num)  # uncomment to apply
print('Outlier analysis complete. Uncomment winsorize_df to cap them.')
```

```python
# ── Prepare feature matrix X and target y ─────────────────────────────────────
FEATURE_COLS = [c for c in df.columns if c != TARGET_COL]

X = df[FEATURE_COLS].copy()
y = df[TARGET_COL].copy()

# Encode target for classification
le = LabelEncoder()
if TASK_TYPE == 'classification':
    y = le.fit_transform(y.astype(str))
    print('Classes:', le.classes_)

# Re-derive column lists from X
num_features = X.select_dtypes(include=np.number).columns.tolist()
cat_features = X.select_dtypes(include=['object', 'category']).columns.tolist()
print(f'Features: {X.shape[1]} | Numeric: {len(num_features)} | Categorical: {len(cat_features)}')
```

```python
# ── Train-Test Split ──────────────────────────────────────────────────────────
TEST_SIZE  = 0.20
stratify   = y if TASK_TYPE == 'classification' else None

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=stratify
)
print(f'Train: {X_train.shape}  |  Test: {X_test.shape}')
```

```python
# ── Full Preprocessing Pipeline ───────────────────────────────────────────────
# Numeric: impute → scale
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler',  StandardScaler()),
])

# Categorical: impute → encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)),
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer,  num_features),
    ('cat', categorical_transformer, cat_features),
], remainder='drop')

# Fit on train, transform both
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep  = preprocessor.transform(X_test)

# Recover feature names
try:
    ohe_names  = preprocessor.named_transformers_['cat']['encoder']\
                              .get_feature_names_out(cat_features).tolist()
except AttributeError:
    ohe_names  = []
feature_names_out = num_features + ohe_names

print(f'Preprocessed shape — Train: {X_train_prep.shape} | Test: {X_test_prep.shape}')
```

```python
# ── Polynomial Features ───────────────────────────────────────────────────────
# poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
# X_poly = poly.fit_transform(X_train_prep)
# poly_names = poly.get_feature_names_out(feature_names_out)
# print(f'Polynomial features: {X_poly.shape[1]}')

# ── Datetime Feature Extraction ───────────────────────────────────────────────
def extract_datetime_features(df: pd.DataFrame, col: str) -> pd.DataFrame:
    df = df.copy()
    df[col] = pd.to_datetime(df[col], errors='coerce')
    df[f'{col}_year']        = df[col].dt.year
    df[f'{col}_month']       = df[col].dt.month
    df[f'{col}_day']         = df[col].dt.day
    df[f'{col}_dayofweek']   = df[col].dt.dayofweek
    df[f'{col}_quarter']     = df[col].dt.quarter
    df[f'{col}_weekofyear']  = df[col].dt.isocalendar().week.astype(int)
    df[f'{col}_is_weekend']  = df[col].dt.dayofweek >= 5
    df[f'{col}_hour']        = df[col].dt.hour
    df.drop(columns=[col], inplace=True)
    return df

# for dcol in date_cols:
#     df = extract_datetime_features(df, dcol)

# ── Log / Sqrt Transforms ─────────────────────────────────────────────────────
def transform_skewed(df: pd.DataFrame, cols: list, threshold: float = 0.75) -> pd.DataFrame:
    df = df.copy()
    for col in cols:
        if abs(df[col].skew()) > threshold:
            if (df[col] > 0).all():
                df[f'{col}_log1p'] = np.log1p(df[col])
            df[f'{col}_sqrt'] = np.sqrt(np.abs(df[col]))
    return df

# df = transform_skewed(df, feat_num)   # uncomment to apply

# ── Interaction Features ──────────────────────────────────────────────────────
def add_interactions(df: pd.DataFrame, col_a: str, col_b: str) -> pd.DataFrame:
    df = df.copy()
    df[f'{col_a}_x_{col_b}']   = df[col_a] * df[col_b]
    df[f'{col_a}_div_{col_b}']  = df[col_a] / (df[col_b] + 1e-9)
    df[f'{col_a}_plus_{col_b}'] = df[col_a] + df[col_b]
    return df

# df = add_interactions(df, 'feature_a', 'feature_b')

print('Feature engineering utilities ready. Uncomment to apply.')
```

```python
# ── Multiple Metrics via cross_validate ──────────────────────────────────────
if TASK_TYPE == 'classification':
    multi_metrics = ['accuracy', 'f1_weighted', 'precision_weighted',
                     'recall_weighted', 'roc_auc_ovr_weighted']
else:
    multi_metrics = ['r2', 'neg_mean_absolute_error',
                     'neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error']

multi_cv = cross_validate(BEST_MODEL, X_train_res, y_train_res,
                          cv=cv, scoring=multi_metrics,
                          return_train_score=True, n_jobs=-1)

print('=== Multi-Metric Cross-Validation ===')
for m in multi_metrics:
    k = f'test_{m}'
    v = multi_cv[k]
    print(f'{m:40s}: {np.abs(v).mean():.4f} ± {np.abs(v).std():.4f}')
```

```python
# ── Final evaluation with tuned model ────────────────────────────────────────
FINAL_MODEL = TUNED_MODEL if PARAM_GRID else MODELS[BEST_MODEL_NAME]
FINAL_MODEL.fit(X_train_res, y_train_res)
y_pred_final = FINAL_MODEL.predict(X_test_prep)

print(f'=== Final Model: {TUNE_MODEL_NAME} ===\n')

if TASK_TYPE == 'classification':
    print(classification_report(y_test, y_pred_final,
                                 target_names=le.classes_.astype(str)))

    # Confusion Matrix
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    ConfusionMatrixDisplay.from_predictions(
        y_test, y_pred_final,
        display_labels=le.classes_,
        ax=axes[0], colorbar=False, cmap='Blues'
    )
    axes[0].set_title('Confusion Matrix')

    # ROC Curve (binary only)
    try:
        y_prob_final = FINAL_MODEL.predict_proba(X_test_prep)
        if n_cls == 2:
            fpr, tpr, _ = roc_curve(y_test, y_prob_final[:, 1])
            auc_val = auc(fpr, tpr)
            axes[1].plot(fpr, tpr, lw=2, label=f'AUC = {auc_val:.4f}')
            axes[1].plot([0, 1], [0, 1], 'k--')
            axes[1].set_xlabel('FPR'); axes[1].set_ylabel('TPR')
            axes[1].set_title('ROC Curve'); axes[1].legend()
        else:
            # Multiclass — one-vs-rest
            for i, cls in enumerate(le.classes_):
                fpr, tpr, _ = roc_curve((y_test == i).astype(int), y_prob_final[:, i])
                axes[1].plot(fpr, tpr, lw=1.5, label=f'{cls} (AUC={auc(fpr,tpr):.2f})')
            axes[1].plot([0, 1], [0, 1], 'k--')
            axes[1].set_title('ROC Curves (OvR)'); axes[1].legend(fontsize=8)
    except Exception as e:
        axes[1].text(0.5, 0.5, f'ROC not available:\n{e}', ha='center', va='center')
    plt.tight_layout(); plt.show()

else:
    r2   = r2_score(y_test, y_pred_final)
    mae  = mean_absolute_error(y_test, y_pred_final)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_final))
    evs  = explained_variance_score(y_test, y_pred_final)
    print(f'R²                   : {r2:.4f}')
    print(f'MAE                  : {mae:.4f}')
    print(f'RMSE                 : {rmse:.4f}')
    print(f'Explained Variance   : {evs:.4f}')

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].scatter(y_test, y_pred_final, alpha=0.4, color='steelblue', s=20)
    mn, mx = min(y_test.min(), y_pred_final.min()), max(y_test.max(), y_pred_final.max())
    axes[0].plot([mn, mx], [mn, mx], 'r--')
    axes[0].set_xlabel('Actual'); axes[0].set_ylabel('Predicted')
    axes[0].set_title(f'Actual vs Predicted  R²={r2:.4f}')

    residuals = y_test - y_pred_final
    sns.histplot(residuals, kde=True, ax=axes[1], color='coral')
    axes[1].axvline(0, color='black', lw=1.5, ls='--')
    axes[1].set_title('Residual Distribution')
    plt.tight_layout(); plt.show()
```

```python
# ── Feature Importance ────────────────────────────────────────────────────────
def plot_feature_importance(model, feature_names, top_n=20, title='Feature Importance'):
    if hasattr(model, 'feature_importances_'):
        imp = model.feature_importances_
    elif hasattr(model, 'coef_'):
        imp = np.abs(model.coef_).flatten()
    else:
        print('Model does not expose feature importances.')
        return
    names = feature_names[:len(imp)]
    fi_df = pd.DataFrame({'feature': names, 'importance': imp})\
              .sort_values('importance', ascending=False).head(top_n)
    plt.figure(figsize=(10, max(5, top_n * 0.35)))
    sns.barplot(data=fi_df, y='feature', x='importance', palette='muted', orient='h')
    plt.title(title); plt.tight_layout(); plt.show()
    return fi_df

fi_df = plot_feature_importance(
    FINAL_MODEL,
    feature_names_out if feature_names_out else [f'f{i}' for i in range(X_train_res.shape[1])]
)
```

## Clustering (Unsupervised)

```python
# ── Elbow + Silhouette for KMeans ─────────────────────────────────────────────
inertias, silhouettes = [], []
K_RANGE = range(2, 11)

for k in K_RANGE:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = km.fit_predict(X_train_prep[:2000])   # sample for speed
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_train_prep[:2000], labels))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(K_RANGE, inertias, 'o-', color='steelblue')
axes[0].set_xlabel('k'); axes[0].set_ylabel('Inertia')
axes[0].set_title('Elbow Method')

axes[1].plot(K_RANGE, silhouettes, 'o-', color='coral')
axes[1].set_xlabel('k'); axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Scores')

plt.tight_layout(); plt.show()
best_k = K_RANGE[np.argmax(silhouettes)]
print(f'Best k by silhouette: {best_k}')
```

```python
# ── Multiple Clustering Algorithms ───────────────────────────────────────────
CLUST_ALGOS = {
    'KMeans'               : KMeans(n_clusters=best_k, random_state=RANDOM_STATE, n_init=10),
    'MiniBatchKMeans'      : MiniBatchKMeans(n_clusters=best_k, random_state=RANDOM_STATE),
    'Agglomerative'        : AgglomerativeClustering(n_clusters=best_k),
    'DBSCAN'               : DBSCAN(eps=0.5, min_samples=5),
    'GaussianMixture'      : GaussianMixture(n_components=best_k, random_state=RANDOM_STATE),
    'Birch'                : Birch(n_clusters=best_k),
    'MeanShift'            : MeanShift(),
    'SpectralClustering'   : SpectralClustering(n_clusters=best_k, random_state=RANDOM_STATE, n_jobs=-1),
}

clust_data = X_train_prep[:1000]
clust_results = []

for name, algo in CLUST_ALGOS.items():
    try:
        labels = algo.fit_predict(clust_data) if hasattr(algo, 'fit_predict') \
                 else algo.fit(clust_data).predict(clust_data)
        n_clust = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clust > 1:
            sil = silhouette_score(clust_data, labels)
            db  = davies_bouldin_score(clust_data, labels)
            ch  = calinski_harabasz_score(clust_data, labels)
        else:
            sil = db = ch = np.nan
        clust_results.append({'Algorithm': name, 'n_clusters': n_clust,
                               'Silhouette': sil, 'Davies-Bouldin': db, 'Calinski-Harabasz': ch})
        print(f'✓ {name:25s} clusters={n_clust}  sil={sil:.3f}  db={db:.3f}')
    except Exception as e:
        print(f'✗ {name}: {e}')

display(pd.DataFrame(clust_results).sort_values('Silhouette', ascending=False))
```
